```python
# rename column iso 3166-1 alpha 3 CODE to Iso_Code
df.rename(columns={'Population':'popu'}, inplace = True)
df
```

```python
# top 5 countries which have more number of covid cases (hint use df.nlargest)
df1= df.nlargest(5, 'Total Cases')
df1[['Country' ,'Total Cases']]
df1
```

```python
#plot bar graph for top 5 countries
sns.barplot(x=df1['Country'],y=df1['Total Cases'])
plt.show()
```

```python
# least 5 countries which have least number of covid cases
df2= df.nsmallest(5, 'Total Cases')
df2[['Country' ,'Total Cases']]
df2
```

```python
#draw bar graph of least 5 countries
sns.barplot(x=df2['Country'],y=df2['Total Cases'])
plt.show()
```

```python
#top 5 countries which have more number of covid Deaths
df3= df.nlargest(5, 'Total Deaths')
df3[['Country' ,'Total Deaths']]
df3
```

```python
#plot bar graph for top 5 countries which have more number of covid Deaths
sns.barplot(x=df3['Country'],y=df3['Total Deaths'])
plt.show()
```

```python
#5 countries which have least number of covid Death
df4= df.nsmallest(5, 'Total Deaths')
df4[['Country' ,'Total Deaths']]
df4
```

```python
#Death percentage
#top 5 countries which have more number of covid  Death percentage 
df5= df.nlargest(5, 'Death percentage')
df5[['Country' ,'Death percentage']]
df5
```

```python
#plot bar graph for top 5 countries which have more number of covid  Death percentage 
sns.barplot(x=df5['Country'],y=df5['Death percentage'])
```

```python
#continent wise cases (hint use group by)
df6=df[["Continent", "Total Cases"]].groupby(['Continent'], as_index=False).sum().sort_values(by='Total Cases', ascending=False)
plt.figure(figsize=(20,8))
sns.barplot(x='Continent', y='Total Cases', data=df6)


```

```python
#draw bar plot for continent wise case
sns.barplot(x='Continent', y='Total Cases', data=df6)
```
