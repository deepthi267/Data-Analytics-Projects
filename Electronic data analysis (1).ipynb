```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("D:\electronics (1).csv")
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df1 =df.dropna()
df1
```

```python
df1.isnull().sum()
```

```python
#step 2:Check for duplicate values and drop it
df1.duplicated().value_counts()
```

```python
df2=df1.drop_duplicates(subset=None, keep ='first',inplace=False)
df2
```

```python
#Step 3
df2.columns=['Order_ID', 'Product', 'Quantity_Ordered','Price_Each','Order_Date','Purchase_Address']

df2.head()
```

```python
#Step 4
df2['Date']=df2['Order_Date'].str[:8]
df2
```

```python
#Step 5
df2['Month']=df2['Order_Date'].str[:2]
df2
```

```python
#step 6
df2[['House_No.','City','Pin_Code']]= df2.Purchase_Address.str.split(",",expand=True)
df2
```

```python
#step 7
df2 = df2[df2['Month'].str[:2] != 'Or']
df2.head()
```

```python
#step 8 correct datatypes
df2['Month'] = df2['Month'].astype('int32')
df2['Order_Date'] = pd.to_datetime(df2['Order_Date'])
df2['Date'] = pd.to_datetime(df2['Date'])
df.info()
```

```python
df2['Quantity_Ordered'] = df2['Quantity_Ordered'].astype('int32')
df2['Price_Each'] = df2['Price_Each'].astype('float64')
df2.info()
```

```python
df2['Sales']=df2['Quantity_Ordered']*df2['Price_Each']
```

```python
#question 3:Sales value by each month
Monthly_Sales = df2.groupby('Month').Sales.sum()
Monthly_Sales
```

```python
#question 4: Which city had the highest sales overall?
city_Sales=df2.City.value_counts()
city_Sales
```

```python
#From the values we can predict clearly that San Francisco giving maximum sale then los angles
#plot the graph

plt.figure(figsize=(16,6))
sns.countplot(x=df2['City'])
plt.xlabel("City Name")
plt.ylabel("Number of Sales")
plt.title("City Vs Sales")
```

```python
#Question 5 Find out the distribution of sales over different times of the day (Morning,
#Afternoon, Evening, Night) and answer the following for each city.
df2['Hour'] = df2.Order_Date.dt.hour
```

```python
df2.groupby('Hour').count()
```

```python
hours = [hour for hour, df2 in df2.groupby('Hour')]
plt.plot(hours, df2.groupby('Hour').count())
plt.grid(True)
plt.xticks(hours)
plt.xlabel('Hours in 24hr format')
plt.ylabel('orders')
plt.show()
```

```python
Highest_orders = df2.Product.value_counts()
Highest_sales = df2.groupby('Product').Sales.sum().sort_values(ascending=False)

#plotting both the charts
fig, axes = plt.subplots(1, 2, figsize=(20,7))

sns.barplot(ax=axes[0], x=Highest_orders.index, y=Highest_orders.values)
axes[0].set_xlabel('Products')
axes[0].set_ylabel('No of Oreders')
axes[0].tick_params(axis='x',rotation=90)

sns.barplot(ax=axes[1],x=Highest_sales.index, y=Highest_sales.values)
axes[1].set_xlabel('Products')
axes[1].set_ylabel('Sales Amount')
axes[1].tick_params(axis='x', rotation=90)
plt.show()
```
